```python
import os 
import glob
import cv2 
import pandas as pd
```

```python


im=[]
count=0
img_dir = "F:/VS Code/Grad k7yan/data"
directory_contents = os.listdir(img_dir)
print(directory_contents)
list_of_dic = []
for d in  directory_contents:
    data_path = os.path.join(img_dir+"/{}".format(d),'*.jpg')
    files = glob.glob(data_path)
    dim=(128,128)
    print(d, len(files))
    i = 1
    for f1 in files:
        dic = {}
        img = cv2.imread(f1)
        resized = cv2.resize(img  , dim , interpolation = cv2.INTER_AREA)
        # f2 = f'F:/VS Code/Grad k7yan/data/{d}/{d}_({i}).jpg'
        # print(f1)
        # print(f2)

        # os.rename(f1, f2)
        # i += 1 
        # print(f1.split('\\')[-1])
        dic['class'] = d
        dic['label'] = f1.split('\\')[-1]
        # im.append(resized)
        list_of_dic.append(dic)
df = pd.DataFrame(list_of_dic)
df
```
